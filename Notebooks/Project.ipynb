#### 7. Define Target and Predictor Variables

#### 8. Vectorize the Text

```python
X = cv.fit_transform(X).toarray()
```

```python
pickle.dump(cv, open('cv.pkl', 'wb'))
```

#### 9. Train Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=4,stratify=y)
```

#### 10. Buliding The Model

##### 10.1 Adding Input Layer 

```python
model.add(Dense(input_dim=5000,kernel_initializer="random_uniform",activation="relu",units=5000))
```

##### 10.2 Adding Hidden Layers

```python
model.add(Dense(units=3000,kernel_initializer="random_uniform",activation="relu"))
```

```python
model.add(Dense(units=1000,kernel_initializer="random_uniform",activation="relu"))
```

##### 10.3 Adding Output Layer

```python
model.add(Dense(units=1,kernel_initializer="random_uniform",activation="sigmoid"))
```

##### 10.4 Compiling the Model

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
```

##### 10.5 Training the Model

```python
model.fit(X_train,y_train,epochs=2,batch_size=32,validation_data=(X_test,y_test))
```

#### 11. Single Predictions

```python
model.predict(cv.transform(['book is amazing']))
```
